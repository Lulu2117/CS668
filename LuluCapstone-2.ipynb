```python
#import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as mdates

# Optional: For clean aesthetics
sns.set(style="whitegrid")
```

### Load the Data

```python
dr = pd.read_excel(r"/Users/louisamoquete/Desktop/CapstoneData/DRData.xlsx")
panama = pd.read_excel(r"/Users/louisamoquete/Desktop/CapstoneData/PanamaData.xlsx")
nicaragua = pd.read_excel(r"/Users/louisamoquete/Desktop/CapstoneData/NicaraguaData.xlsx")
dr['Year'] = pd.to_datetime(dr['Year'], format='%Y')
panama['Year'] = pd.to_datetime(panama['Year'], format='%Y')
nicaragua['Year'] = pd.to_datetime(nicaragua['Year'], format='%Y')
```

### GDP Line Plots

```python
# Plot GDP for Dominican Republic with highlight
plt.figure(figsize=(10, 6))
sns.lineplot(data=dr, x='Year', y='GDP', color='midnightblue')
plt.title('GDP Over Time  Dominican Republic (with U.S. Occupation Highlighted)')
plt.ylabel('GDP (in current USD or local currency)')
plt.xlabel('Year')

# Highlight 1965 to 1966
plt.axvspan(pd.to_datetime("1965", format="%Y"),
            pd.to_datetime("1966", format="%Y"),
            color='lightskyblue', alpha=0.3, label='U.S. Occupation (19651966)')

plt.legend()
plt.show()

# Plot GDP for Panama with U.S. invasion highlight
plt.figure(figsize=(10, 6))
sns.lineplot(data=panama, x='Year', y='GDP', color='midnightblue')
plt.title('GDP Over Time  Panama (with U.S. Invasion Highlighted)')
plt.ylabel('GDP (in current USD or local currency)')
plt.xlabel('Year')

# Highlight U.S. invasion (19891990)
plt.axvspan(pd.to_datetime("1989-01-01"), pd.to_datetime("1990-12-31"),
            color='tan', alpha=0.3, label='U.S. Invasion (19891990)')

plt.legend()
plt.show()

# Plot GDP for Nicaragua with U.S. intervention highlight
plt.figure(figsize=(10, 6))
sns.lineplot(data=nicaragua, x='Year', y='GDP', color='midnightblue')
plt.title('GDP Over Time  Nicaragua (with U.S. Intervention Highlighted)')
plt.ylabel('GDP (in current USD or local currency)')
plt.xlabel('Year')

# Highlight U.S. involvement (19811990)
plt.axvspan(pd.to_datetime("1981-01-01"), pd.to_datetime("1990-12-31"),
            color='thistle', alpha=0.3, label='U.S. Involvement (19811990)')

plt.legend()
plt.show()
```

## Year-Over-Year % Change

```python
# Calculate YoY GDP change
dr['GDP_YoY_Change'] = dr['GDP'].pct_change() * 100

# Plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=dr, x='Year', y='GDP_YoY_Change', color='midnightblue')
plt.title('YoY GDP Growth  Dominican Republic (with U.S. Occupation Highlighted)')
plt.ylabel('% Change')
plt.xlabel('Year')

# Highlight U.S. occupation period (19651966)
plt.axvspan(pd.to_datetime("1965-01-01"), pd.to_datetime("1966-12-31"),
            color='thistle', alpha=0.3, label='U.S. Occupation (19651966)')

plt.legend()
plt.show()
```

## Correlation Heatmaps

```python
#GDP vs. FDI  Panama
fig, ax1 = plt.subplots(figsize=(10,6))

# GDP line
sns.lineplot(data=panama, x='Year', y='GDP', ax=ax1, label='GDP', color='midnightblue')
ax2 = ax1.twinx()

# FDI line on right Y-axis
sns.lineplot(data=panama, x='Year', y='FDI', ax=ax2, label='FDI', color='darkgreen')
plt.title('GDP vs. FDI  Panama')


#GDP vs. FDI  Panama with U.S. occupation highlight
fig, ax1 = plt.subplots(figsize=(10,6))

# GDP line (left Y-axis)
sns.lineplot(data=panama, x='Year', y='GDP', ax=ax1, label='GDP', color='midnightblue')
ax1.set_ylabel('GDP', color='midnightblue')

# Highlight U.S. occupation period on ax1
ax1.axvspan(pd.to_datetime("1989-01-01"), pd.to_datetime("1990-12-31"),
            color='tan', alpha=0.3, label='U.S. Invasion (19891990)')

# FDI line (right Y-axis)
ax2 = ax1.twinx()
sns.lineplot(data=panama, x='Year', y='FDI', ax=ax2, label='FDI', color='darkgreen')
ax2.set_ylabel('FDI', color='darkgreen')

# Title and combined legend
plt.title('GDP vs. FDI  Panama (with U.S. Invasion Highlighted)')
fig.legend(loc="upper left", bbox_to_anchor=(0.1, 0.9))

plt.show()
```

```python
# GDP vs. Inflation  Panama with U.S. occupation highlight
fig, ax1 = plt.subplots(figsize=(10, 6))

# GDP line (left Y-axis)
sns.lineplot(data=panama, x='Year', y='GDP', ax=ax1, label='GDP', color='midnightblue')
ax1.set_ylabel('GDP', color='midnightblue')

# Highlight U.S. invasion (19891990)
ax1.axvspan(pd.to_datetime("1989-01-01"), pd.to_datetime("1990-12-31"),
            color='thistle', alpha=0.3, label='U.S. Invasion (19891990)')

# Inflation line (right Y-axis)
ax2 = ax1.twinx()
sns.lineplot(data=panama, x='Year', y='Inflation', ax=ax2, label='Inflation', color='maroon')
ax2.set_ylabel('Inflation Rate (%)', color='maroon')

# Title and combined legend
plt.title('GDP vs. Inflation  Panama (with U.S. Invasion Highlighted)')
fig.legend(loc="upper left", bbox_to_anchor=(0.1, 0.9))

plt.show()
```
